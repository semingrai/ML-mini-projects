```python
import pandas as pd
data = pd.read_csv('ecommerce_sales_data.csv')
data.head(2)
```

```python
data.columns
```

```python
features = ['Order Date', 'Product Name', 'Category', 'Region', 'Quantity', 'Sales']
X = data[features]
Y = data.Profit
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, Y_train, Y_test = train_test_split(X,Y, random_state = 1)
s = (X_train.dtypes == 'object')
obj_cols = list(s[s].index)
```

```python
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import mean_absolute_error
def score_data(X_train, X_test, Y_train, Y_test):
    model = RandomForestRegressor(random_state = 1)
    model.fit(X_train, Y_train)
    pred = model.predict(X_test)
    return mean_absolute_error(Y_test, pred)
```

```python
from sklearn.preprocessing import OneHotEncoder
OH = OneHotEncoder(handle_unknown='ignore', sparse_output=False)
OH_X_train = pd.DataFrame(OH.fit_transform(X_train[obj_cols]))
OH_X_test = pd.DataFrame(OH.transform(X_test[obj_cols]))
OH_X_train.index = X_train.index
OH_X_test.index = X_test.index
```

```python
w_X_train = X_train.drop(obj_cols, axis = 1)
w_X_test = X_test.drop(obj_cols, axis = 1)
```

```python
OH_X_train = pd.concat([OH_X_train, w_X_train], axis = 1)
OH_X_test = pd.concat([OH_X_test, w_X_test], axis=1)
```

```python
OH_X_train.columns = OH_X_train.columns.astype('str')
OH_X_test.columns = OH_X_test.columns.astype('str')
```

```python
mae = score_data(OH_X_train, OH_X_test, Y_train, Y_test)
score = Y.mean()
error = (score/mae) * 100

print(f"{mae}\n {score}\n {error}")
```
