```python
import pandas as pd
data = pd.read_csv('Engineering_graduate_salary.csv')
data.head(2)
```

```python
data.columns
```

```python
features = ['    ID', 'Gender', 'DOB', '10percentage', '10board', '12graduation',
       '12percentage', '12board', 'CollegeID', 'CollegeTier', 'Degree',
       'Specialization', 'collegeGPA', 'CollegeCityID', 'CollegeCityTier',
       'CollegeState', 'GraduationYear', 'English', 'Logical', 'Quant',
       'Domain', 'ComputerProgramming', 'ElectronicsAndSemicon',
       'ComputerScience', 'MechanicalEngg', 'ElectricalEngg', 'TelecomEngg',
       'CivilEngg', 'conscientiousness', 'agreeableness', 'extraversion',
       'nueroticism', 'openess_to_experience']
Y = data.Salary
X = data[features]
```

```python
from sklearn.model_selection import train_test_split
train_X, val_X, train_y, val_y = train_test_split(X, Y, train_size=0.8, test_size=0.2, random_state= 0)
```

```python
categorical = [i for i in train_X.columns if train_X[i].nunique()<20 and train_X[i].dtype == 'object']
numerical = [i for i in train_X.columns if train_X[i].dtype in ['float64', 'int64']]
```

```python
my_cols = categorical+numerical
X_train = train_X[my_cols].copy()
X_val = val_X[my_cols].copy()
```

```python
from sklearn.compose import ColumnTransformer
from sklearn.pipeline import Pipeline
from sklearn.impute import SimpleImputer
from sklearn.preprocessing import OneHotEncoder

numerical_transformer = SimpleImputer(strategy='constant')
categorical_transformer = Pipeline(steps=[
    ('imp', SimpleImputer(strategy='most_frequent')),
    ('hot', OneHotEncoder(handle_unknown='ignore'))
])

preprocessor = ColumnTransformer(transformers=[
    ('num', numerical_transformer, numerical),
    ('cat', categorical_transformer, categorical)
])
```

```python
from sklearn.ensemble import RandomForestRegressor
model = RandomForestRegressor(n_estimators=50,random_state=0)
```

```python
from sklearn.metrics import mean_absolute_error
my_pipeline = Pipeline(steps=[
    ('pre', preprocessor),
    ('model', model)
])
my_pipeline.fit(X_train, train_y)
pred = my_pipeline.predict(val_X)
```

```python
wai = Y.mean()
mae = mean_absolute_error(val_y, pred)
error = (mae/wai)*100
print(f"avg: {wai}")
print(f"mae: {mae}")
print(f"error:{error}")
```
